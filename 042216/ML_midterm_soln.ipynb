# Problem 1

Let $R = R(\theta)$ be a rotation matrix with a rotational angle of $\theta$ in $\mathbb{R}^n$.

1. Prove that $$R^TR=I \;\text{in}\; \mathbb{R}^n$$
    
    Hint: Euclidian distances of vectors are preserved after a rotational operation. (_i.e._, $\lVert Rx \rVert = \lVert x \rVert$ for any $x \in \mathbb{R}^n$ ) 
<br><br>
2. Prove that $$R^T(\theta) = R^{-1}(\theta) = R(-\theta)\;\text{in}\;\mathbb{R}^n$$
<br><br>
3. Show that column vectors in $R$ are orthogonal in $\mathbb{R}^n$.

## Solution

1. $\lVert Rx \rVert ^2 = (Rx)^T(Rx) = x^TR^TRx = \lVert x \rVert ^2 = x^Tx\,$ for any $x \implies R^TR = I$

2. $R^TR = I \implies R^{-1} = R^T$

3. $R^TR = I \implies r_i^T r_j = \delta_{ij}$

# Problem 3

Find solution $D$ of the optimization problem with the following objective function and constraints. Here, $\lVert \cdot \rVert_2$ is the Euclidean norm (or distance). 

<br><br>
$$\begin{align*} D= \min_Z \quad
& \lVert X_1 - Z \rVert_2 + \lVert X_2 - Z \rVert_2 \\ \\
\text{subject to} \quad
& Z \in V=\text{span} \{V_1, V_2\}
\end{align*}$$

<img src="./image_files/mirrored_line.png", width = 300>
<center>Figure 2</center>

## Solution

$D^*$ is the shortest distance from $X_1$ to $X_2$ via $Z$ on the spanned hyperplane. The shortest line is the straight line from $X_1$ and $Y$ which is mirrored reflection of $X_2$ with respect to the hyperplane.

$$
\begin{align*}
D^* &= \lVert X_1 - MX_2 \rVert = \lVert X_1 - \left( 2V \left(V^TV \right)^{-1}V^T-I\right)X_2\rVert \\
&=\lVert MX_1 - X_2\rVert
\end{align*}
$$
where $M$ is mirror transformation and $V = [V_1 \; V_2]$

# Problem 5

The regularized least-squares problem has the form

<br>
$$ \min_{\theta} \;\lVert A\theta -y\rVert_2^2 + \lambda \lVert \theta \rVert_2^2$$

1. Show that the solution is given by
<br><br>
$$ \hat{\theta} = \left( A^T A + \lambda I_n \right)^{-1} A^T y $$
 (Do not use the method of Lagrangian multipliers)
<br><br>
2. Find and draw an approximated curve of the given data points in either Matlab or python
    - overcome overfitting 
    - use 10 RBFs with $\sigma = 1$
    - choose a proper value of $\lambda$ on your own

## Solution

$$
\begin{align*} \left\Vert A\theta - y\big\Vert^2_{2} + \big\Vert \sqrt{\lambda}I\theta-0\right\Vert^2_{2}
& = \Bigg\Vert 
\begin{bmatrix}
A\\
\sqrt{\lambda}I
\end{bmatrix}
\theta-
\begin{bmatrix}
y\\
0
\end{bmatrix}
\Bigg\Vert^2_{2}
\end{align*}$$

$$ 
\hat{\theta} = \left( \begin{bmatrix}
A\\
\sqrt{\lambda}I
\end{bmatrix}^T \begin{bmatrix}
A\\
\sqrt{\lambda}I
\end{bmatrix}\right)^{-1}\begin{bmatrix}
A\\
\sqrt{\lambda}I
\end{bmatrix}^T
\begin{bmatrix}
y\\
0
\end{bmatrix} = \left( A^T A + \lambda I_n \right)^{-1} A^T y
$$

```octave
x = -4.5:1:4.5; 
x = x(:);

y = [0.9819 0.7973 1.9737 0.1838 1.3180 -0.8361 -0.6591 -2.4701 -2.8122 -6.2512]';

xpt = -4.5:0.01:4.5; 
xpt = xpt(:);

d = 10;
u = linspace(-4.5,4.5,d);

sigma = 1;

A = [];
rbfbasis = [];
for i = 1:d
    A = [A exp(-(x-u(i)).^2/(2*sigma^2))];
    rbfbasis = [rbfbasis exp(-(xpt-u(i)).^2/(2*sigma^2))];    
end

lambda = 0.1;
theta = inv(A'*A + lambda*eye(d))*A'*y;

ypt = rbfbasis*theta;
plot(x,y,'o',xpt,ypt)
axis([-5 5 -12 6])
```

PCA can be used for dimensionality reduction. We want to explore this with the following data set: two variables (‘data’ and ‘group’) in the workspace.
```octave
%%
m = 200;
X = data;

Xbar = mean(X,1);
X = X - repmat(Xbar,m,1);

S = cov(X);

[V,D] = eig(S);

%%
% sort in decreasing order
lambda = diag(D);
[dummy,m_arrange] = sort(-1*lambda);
lambda = lambda(m_arrange);
V = V(:,m_arrange); 

figure(1),  clf
stem(sqrt(lambda),'filled'),  xlim([0.8 5.2])
```

```octave
%%
Y = X*V;

X1 = Y(:,1);
X2 = Y(:,2);
C1 = 1:100;
C2 = 101:200;

plot(X1(C1),X2(C1),'go',X1(C2),X2(C2),'bo')
```

```octave
%% SVM
A = [X1(C1)';
    X2(C1)'];

B = [X1(C2)';
    X2(C2)'];


% g = 0.1;
g = 1;
% g = 10;

N = 100;
M = 100;

cvx_begin quiet
    variables w(2) w0 u(N) v(M)
    minimize (norm(w,2) + g*(ones(1,N)*u + ones(1,M)*v))
    subject to
        A'*w + w0 >= 1 - u;
        B'*w + w0 <= -(1 - v);
        u >= 0;
        v >= 0;
cvx_end

t = -8:0.1:8;
p = -w(1)/w(2)*t - w0/w(2);
p1 = p + 1/w(2);
p2 = p - 1/w(2);

figure(6)
h6 = plot(X1(C1),X2(C1),'go',X1(C2),X2(C2),'bo',...  
    t,p,'r--',...
    t,p1,'k',...
    t,p2,'k');

axis equal
xlim([-8 8])

xlabel('x_1'),   ylabel('x_2')
title('Classification using SVM','fontsize',10);
legend('C1','C2')
```

```octave
%% preceptron

X3 = [A B];
y1 = ones(1,length(C1));    % class 1 => y = +1
y2 = -ones(1,length(C2));   % class 2 => y = ?1
y = [y1 y2];
X3 = [ones(1,length(X3));X3];


w = ones(3,1);  % w initialization
iter = 100;      % iteration number
for k = 1:iter
    for j = 1:iter
        i = randi(length(X3),1);
        yhat = sign(w'*X3(:,i));
        w = w + 1/2*(y(i)-yhat)*X3(:,i);
    end
end

% to draw decision boundary
t_min = min(X3(2,:));
t_max = max(X3(2,:));
t = linspace(t_min,t_max,100);
p = -w(2)/w(3)*t - w(1)/w(3);

h2 = plot(X1(C1),X2(C1),'go',X1(C2),X2(C2),'bo',t,p,'--r');
axis equal
xlim([-8 8])
xlabel('x_1'), ylabel('x_2')
legend('C1','C2','perceptron','location','NorthEast')
title('Classification using Perceptron','fontsize',10);
```
