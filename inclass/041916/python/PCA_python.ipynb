```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA
%pylab inline
```

```python
#data generation

m = 1000
mu = np.array([0, 0])
sigma = np.array([[1, 1.5], [1.5, 3]])
r = np.random.multivariate_normal(mu, sigma, m) 
x = np.transpose(np.array([r[:, 0]]))
y = np.transpose(np.array([r[:, 1]]))

X = np.concatenate((x, y), axis = 1)
np.shape(r)
```

```python
#Assume we do not know mu and SIGMA only X is given
S = np.cov(np.transpose(r))
V, D = LA.eig(S)

V = np.diag(V)
u1 = D[:, 1]/(LA.norm(D[:, 1])) 
h = u1[1]/u1[0]

plt.plot(X[:, 0], X[:, 1], 'go')
plt.plot(x, x.dot(h), 'r')
axis('equal')
plt.show()
```

```python
Z = r.dot(u1) # this is new 1dim (reduced or porjected) data vis PCA
plt.hist(Z, 21)
plt.show()
```
