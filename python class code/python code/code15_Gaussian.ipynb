```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (8, 8)

theta = np.pi/3

R = np.array([[np.cos(theta), -np.sin(theta)],
     [np.sin(theta), np.cos(theta)]])

z = R.dot(y);
print np.size(z)
mu = np.array([[3], [0]]);
w = z + np.repeat(mu,z.shape[1], axis = 1);


plt.plot(w[0,:],w[1,:],'.')
plt.axis('equal')
plt.xlim([-10., 10.])
```

```python
m = w.shape[1];
mu_hat = np.array([np.mean(w, 1)]).transpose()
mu_hat
```

```python
z = w - np.repeat(mu_hat,m, axis = 1)
S = 1./m*z.dot(z.transpose())
S
```

```python
[D, U] = LA.eig(S)
# sorting
sort_indices = np.argsort(D)[::-1]
D[:] = D[sort_indices]
U = U[:,sort_indices]
print (D)
print (U)
```
