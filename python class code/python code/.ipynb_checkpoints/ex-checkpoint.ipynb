```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
from cvxpy import *
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# data points in column vector input, output
x = np.array([[0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]]).transpose()
y = np.array([[0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]]).transpose()

m = y.shape[0]
A = np.concatenate((x,np.ones((m,1))),1)

#theta = LA.inv(A.transpose().dot(A)).dot(A.transpose()).dot(y)
```

```python
#only for python2
#optimiza theta  use norm 1

theta1 = Variable(A.shape[1])
objective = Minimize(cvxpy.norm1(A*theta1 - y))
constraints = []
prob = Problem(objective, constraints)
result = prob.solve()
print(theta1.value)

#to plot
plt.plot(x, y, 'ko')
xp1 = np.linspace(0, 5, 500)
yp1 = theta1[0].value*xp1 + theta1[1].value


plt.plot(xp1, yp1, 'r-', linewidth = 2)

plt.axis('equal')
plt.legend(['data', 'L_1'], loc = 4)
plt.show()
```

```python
#only for python2
#optimiza theta  use norm 1

theta2 = Variable(A.shape[1])
objective = Minimize(cvxpy.norm2(A*theta2 - y))
constraints = []
prob = Problem(objective, constraints)
result = prob.solve()
print(theta2.value)

#to plot
plt.plot(x, y, 'ko')
xp2 = np.linspace(0, 5, 500)
yp2 = theta2[0].value*xp2 + theta2[1].value


plt.plot(xp2, yp2, 'r-', linewidth = 2)

plt.axis('equal')
plt.legend(['data', 'L_2'], loc = 4)
plt.show()
```

```python
x = np.array([[0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]]).transpose()
y = np.array([[0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]]).transpose()

#add outliers
x = np.concatenate((x,np.array([[0.5], [3.8]])))
y = np.concatenate((y,np.array([[3.9], [0.3]])))

m = y.shape[0]
A = np.concatenate((x,np.ones((m,1))),1)

#use norm1

theta1 = Variable(A.shape[1])
objective = Minimize(cvxpy.norm1(A*theta1 - y))
constraints = []
prob = Problem(objective, constraints)
result = prob.solve()

#to plot
plt.plot(x, y, 'ko')
xp1 = np.linspace(0, 5, 500)
yp1 = theta1[0].value*xp1 + theta1[1].value


#use norm2

theta2 = Variable(A.shape[1])
objective = Minimize(cvxpy.norm2(A*theta2 - y))
constraints = []
prob = Problem(objective, constraints)
result = prob.solve()


plt.plot(xp1, yp1, 'b-', linewidth = 2)

xp2 = np.linspace(0, 5, 500)
yp2 = theta2[0].value*xp2 + theta2[1].value


plt.plot(xp2, yp2, 'r-', linewidth = 2)
plt.axis('equal')
plt.legend(['data', 'L_1', 'L_2'], loc = 4)
plt.show()
```

```python
# Data Generation
from mpl_toolkits.mplot3d import Axes3D

n = 500

x1 = np.random.randn(n,1)
x2 = np.random.randn(n,1)

real_theta1 = 1;
real_theta2 = 3;
real_theta3 = 2;

e = 0.5*np.random.randn(n,1)
y = real_theta1*x1 + real_theta2*x2 + real_theta3 + e

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y)


A = np.concatenate((x1, x2, np.ones((n,1))), 1)
theta = LA.inv(A.transpose().dot(A)).dot(A.transpose()).dot(y)

[X1, X2] = np.meshgrid(np.linspace(np.min(x1), np.max(x1), 25),
                      np.linspace(np.min(x2), np.max(x2), 25))
yhat = theta[0]*X1 + theta[1]*X2 + theta[2]


ax.plot_wireframe(X1, X2, yhat)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

#use sklearn library
model = LinearRegression(normalize=True)

x = np.array([[0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]]).transpose()
y = np.array([[0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]]).transpose()

model.fit(x, y)

plt.plot(x, y, 'ko')
xp3 = np.linspace(0, 5, 500)
yp3 = model.coef_[0]*xp2 + model.intercept_


plt.plot(xp3, yp3, 'g-', linewidth = 2)

plt.axis('equal')
plt.legend(['data', 'L'], loc = 4)
plt.show()


```
