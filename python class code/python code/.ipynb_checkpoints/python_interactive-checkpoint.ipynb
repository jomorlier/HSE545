```python
from numpy import linalg as LA
from numpy import matlib as matlib
import numpy as np

import matplotlib.pyplot as plt
%pylab inline
```

# De nosing

```python
#generate a corrupted time signal
n = 1000
t = np.transpose(np.array(range(0, n, 1)))
exact = 0.5*np.sin((2*np.pi/n)*t)*(np.sin(0.01*(t)))
corrupt = exact + 0.05*np.random.randn(np.size(exact))



def plot_denoise(mu):
    D = np.zeros((n-1,n))
    D[:, 0:n-1] = -np.eye(n-1)
    D[:, 1:n] = D[:, 1:n] + np.eye(n-1)

    corrupt_mtx = np.transpose((np.array([corrupt]))) 
    A = np.concatenate((np.eye(n), np.sqrt(mu)*D), axis = 0)
    b = np.concatenate((corrupt_mtx, np.zeros((n-1, 1))), axis = 0)
    sol = LA.inv(np.transpose(A).dot(A)).dot(np.transpose(A)).dot(b)
    
    plt.plot(t, corrupt, '-', t, sol, 'r', linewidth = 2)
    plt.show()
    
from IPython.html.widgets import interact
interact(plot_denoise, mu = [0, 10000, 500]);
```

# fisher

```python
#training data set generation

n0 = 200;
n1 = 200

mu = np.array([0, 0])
sigma = np.array([[0.9, -0.4],[-0.4, 0.3]])

x0 = np.transpose(np.random.multivariate_normal(mu, sigma, n0)) + 2.5 * np.ones([2,n0])  # data in class 0
x1 = np.transpose(np.random.multivariate_normal(mu, sigma, n1)) + 1 * np.ones([2,n1])# data in class 1
mu0 = np.mean(x0, axis = 1)
mu1 = np.mean(x1, axis = 1)
s0 = 1/(n0-1)*(x0-matlib.repmat(mu0,2,n0/2)).dot(np.transpose(x0-matlib.repmat(mu0,2,n0/2)))
s1 = 1/(n1-1)*(x1-matlib.repmat(mu1,2,n1/2)).dot(np.transpose(x1-matlib.repmat(mu1,2,n1/2)))
```

```python
#make plot
#different w (x axis)
def plot_diff_w1(w_X = 10, w_Y= 10):
    w = np.array([[w_X], [w_Y]])
    
    plt.subplot(221)
    plt.scatter(x0[0,:],x0[1,:], c = 'red');
    plt.scatter(x1[0,:],x1[1,:], c = 'blue')
    r = np.array([[x/10 for x in range(-20,61)]])
    ln = (w/LA.norm(w)).dot(r)
    plt.plot(ln[0,:],ln[1,:],'k',linewidth = 3)
    
    plt.subplot(222)
    y1 = np.transpose(x0).dot(w)
    y2 = np.transpose(x1).dot(w)
    plt.hist((y1,y2),20)
    
    plt.show()

from IPython.html.widgets import interact
interact(plot_diff_w1, w_X=[0, 10, 1], w_Y=[0, 10, 1]);

```

```python
def plot_diff_w2(X_input):
    w_opt = LA.inv(n0*s0+n1*s1).dot(mu0-mu1)
    w_opt = np.transpose(np.array([w_opt]))
    w_category = [[[0], [10]], [[10], [0]], [[5], [5]], [[1], [3]], w_opt]
    X = w_category[X_input]
    w = np.array(X)
    plt.subplot(221)
    plt.scatter(x0[0,:],x0[1,:], c = 'red');
    plt.scatter(x1[0,:],x1[1,:], c = 'blue')
    r = np.array([[x/10 for x in range(-20,61)]])
    ln = (w/LA.norm(w)).dot(r)
    plt.plot(ln[0,:],ln[1,:],'k',linewidth = 3)
    
    plt.subplot(222)
    y1 = np.transpose(x0).dot(w)
    y2 = np.transpose(x1).dot(w)
    plt.hist((y1,y2), 20)
    
    plt.show()


from IPython.html.widgets import interact

interact(plot_diff_w2, X_input = (0, 4, 1));
```

# Statistics

## Sample Mean and Sample Size
- True for any probability density functions

```python
# various sample size m
def plot_large_num(sample_size):
    m = np.array([[x for x in range(10,sample_size,10)]])
    means = np.zeros((1, m.shape[1])) 
    for i in range(0, np.shape(m)[1]):
        x = np.random.normal(10, 30, m[0, i])
        means[0, i] = np.mean(x)

    plt.plot(m, means, 'bo', markersize = 4)
    plt.plot([0, 1000], [10, 10], 'k--')
    plt.axis([0,1000,-10,30])
    plt.xlabel('# of smaples (= sample size)', fontsize = 10)
    plt.ylabel('sample mean', fontsize = 10)
    plt.show()
    
from IPython.html.widgets import interact
interact(plot_large_num, sample_size=[0, 1000, 10]);
```

## Variance Gets Smaller as m is Larger
- Seems approximately Gaussian distributed 
- numerically demostrate that sample mean follows the Gaussin dist

```python
## plot - s 800, 420
N = 100

# sample of size m

def plot_Central_limit(m_size):
    # sample mean (or sample average)
    S = np.array([])
    S = np.concatenate((S, np.mean(np.random.rand(m_size, m_size), axis = 0)))

    plt.hist(S, 21)
    plt.xlim([0, 1])
    title = 'm = %s' % str(m_size)  
    plt.title(title)
    plt.show()
from IPython.html.widgets import interact
interact(plot_Central_limit, m_size=[10, 2000, 10]);
```
