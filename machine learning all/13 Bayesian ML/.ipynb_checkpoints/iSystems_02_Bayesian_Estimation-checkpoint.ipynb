- $H$: Hypothesis, hidden state
- $D = \{d_1,d_2,\cdots,d_m\}$: data, observation, evidence

$$ P(H,D) = P(H \mid D)P(D) = P(D \mid H)P(H) $$

$$ P(H \mid D) = \frac{P(D \mid H)P(H)}{P(D)} \;\;\; \text{: Bayes' Rule} $$

## Combining Evidence

$$\begin{align*} P(H \mid d_1,d_2,\cdots,d_m) &= \frac{P(d_1,d_2,\cdots,d_m \mid H)P(H)}{P(d_1,d_2,\cdot,d_m)}  \\
&=\frac{P(d_1 \mid H)P(d_2 \mid H)\cdots P(d_m \mid H)P(H)}{P(d_1,d_2,\cdots,d_m)} \\ 
& \because \; \text{conditional independence (Markovian Property)}
\end{align*}$$

### Note

$$ \begin{align*} P(H \mid d_1,d_2,\cdots,d_m) &= \frac{P(d_m \mid H,d_1,\cdots ,d_{m-1})P(H \mid d_1,d_2,\cdots,d_{m-1})}{P(d_m \mid d_1,d_2,\cdots,d_{m-1})}\\
&= \eta\prod\limits_{i=1}^m P(d_i \mid H)P(H) \end{align*} $$

- When multiple $d_1, d_2, \cdots $

$$ P(H \mid d_1,d_2,\cdots,d_m) = \frac{P(d_1 \mid H)P(d_2 \mid H)\cdots P(d_m \mid H)P(H)}{P(d_1,d_2,\cdots,d_m)} $$

<center>OR</center>

$$P_0(H) \implies \; P(H \mid d_1)=P_1(H) \implies \; P(H \mid d_2) =  P_2(H) \implies \; \cdots$$

- Recursive Bayesian Estimation

$$ P(H \mid D) = \frac{P(D \mid H)P(H)}{P(D)} $$

<center>posterior $\rightarrow$ prior <br> as more evidence is collected</center>

### given model

```octave
mu = 50;        % if location of unknown object = 50
sigma = 40;     % assume normal dist w/ sigma = 40

% plot its pdf
n = 100;
xt = linspace(1,n)';
figure(1),  clf
plot(xt,normpdf(xt,mu,sigma))
xlabel('D')
ylabel('pdf')
```

```octave
pr = unidpdf(1:n,n)';

M = zeros(n,n);
for i = 1:n
    H = xt(i);
    M(:,i) = normpdf(xt,H,sigma);
end

figure(3),  clf
imagesc(M), axis image
set(gca,'ydir','normal')
xlabel('H'),    ylabel('D'),    title('Pr(D|H)')
```

### likelihood, $P(D \mid H)$

```octave
H = xt;
D = xt;

[X1,X2] = meshgrid(H,D);

F = normpdf(X2,X1,sigma); 

figure(7),  clf
h1 = plot(D,F(:,1));
title('Likelihood, Pr(D|H)')
xlabel('D')
ht = text(40,0.008,['H = ',num2str(1)]);
set(ht,'fontsize',14)
ylim([0 0.014])
    
for i = 1:n    
    set(h1,'YData',F(:,i))    
    set(ht,'String',['H = ',num2str(i)])
    drawnow
    pause(0.01)
end
```

$$P(D \mid H) P(H) \text{ and }P(H \mid D) = \frac{P(D \mid H) P(H)}{P(D)}$$
- suppose d = 30

```octave
d = 30;     % obervation
pr1 = unidpdf(1:n,n)';
pr2 = normpdf(xt,60,sigma); % suppose true location = 40

% normalize
po1 = F(:,d).*pr1;
po2 = F(:,d).*pr2;
po1 = po1/sum(po1);
po2 = po2/sum(po2);


figure(6),  clf
subplot(3,1,1),  plot(D,F(:,d).*pr1,'r',D,F(:,d).*pr2,'b')
legend('P(D|H_1)P(H_1)','P(D|H_2)P(H_2)')
title('Effect of Different Prior')

subplot(3,1,2),  plot(D,po1,'r',D,pr1,'r--')
legend('P(H_1|D)','P(H_1)')
ylabel('probability')

subplot(3,1,3),  plot(D,po2,'b',D,pr2,'b--')
legend('P(H_2|D)','P(H_2)')
xlabel('D')
ylabel('probability')
```

```octave
surf(H,D,F),    hold on
shading interp
view(58,68)
xlabel('H')
ylabel('D')

h32 = plot3(1*ones(n,1),D,F(:,1),'r'); 
h33 = plot3(H,1*ones(n,1),F(1,:),'r'); hold off

% break
for i = 1:n
    set(h32,'XData',i*ones(n,1),'ZData',F(:,i))
    set(h32,'linewidth',2)
    drawnow
    pause(0.01)    
end

for i = 1:n
    set(h33,'YData',i*ones(n,1),'ZData',F(i,:))
    set(h33,'linewidth',2)
    drawnow
    pause(0.01)    
end
```

```octave
subplot(2,2,3)
surf(H,D,F),    hold on
shading interp
view(58,68)
xlabel('H')
ylabel('D')

h32 = plot3(1*ones(n,1),D,F(:,1),'r'); 
h33 = plot3(H,1*ones(n,1),F(1,:),'r'); hold off

subplot(2,2,1)
h1 = plot3(1*ones(n,1),D,F(:,1),'r','linewidth',2); 
view(58,68),    grid on
xlabel('H')
ylabel('D')
title('P(D|H)')

subplot(2,2,4)
h2 = plot3(H,1*ones(n,1),F(1,:),'r','linewidth',2); 
view(58,68),    grid on
xlabel('H')
ylabel('D')
title('P(H|D)')

for i = 1:n
    set(h1,'XData',i*ones(n,1),'ZData',F(:,i))
    set(gca,'xtick',i)
    set(gca,'xticklabel',i)
    set(h32,'XData',i*ones(n,1),'ZData',F(:,i))
    set(h32,'linewidth',2)
    drawnow
    pause(0.04)    
end

for i = 1:n
    set(h2,'YData',i*ones(n,1),'ZData',F(i,:))
    set(h33,'YData',i*ones(n,1),'ZData',F(i,:))
    set(h33,'linewidth',2)
    drawnow
    pause(0.01)    
end
```

```octave
sigma = 10;

figure(10); clf 

pr = unidpdf(1:n,n)';
d = normrnd(mu,sigma);    
D = round(d);           % observation

if D <= 0
    D = 1;
elseif D >= 100
    D = 99;
end

tmp = F(:,D).*pr;   % P(D|H)P(H)
po = tmp/sum(tmp);  % P(H|D)

pr = po;            % recursive

h = plot(H,pr);
ht2 = text(20,0.02,['D = ',num2str(D)]);
set(ht2,'fontsize',14)
xlabel('H')
ylabel('P(H|D)=P(H|d_1d_2...d_m)')

Drecord = [];
for i = 1:30
   
    d = normrnd(mu,sigma);    
    D = round(d);           % observation
    
    if D <= 0
        D = 1;
    elseif D >= 100
        D = 99;
    end
    Drecord = [Drecord; D];
    
    tmp = F(:,D).*pr;   % P(D|H)P(H)
    po = tmp/sum(tmp);  % P(H|D)
    
    set(h,'ydata',pr)
    set(ht2,'string',['D = ',num2str(D)])
    
    pr = po;            % recursive
    pause(0.1)
end
hold off
[Y,I] = max(pr);
I
Drecord
```

```octave
hist(Drecord),  xlim([0,100])
title('Histogram of D')
```
