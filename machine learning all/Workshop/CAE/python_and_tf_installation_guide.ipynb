# 1. Overview 


- 내용
    - 본 매뉴얼은 __기계학습__ 및 __딥러닝__을 위한 패키지 설치를 다룸
    - 실습은 패키지가 이미 설치되어 있는 전산실에서 진행되므로 개인적으로 설치하지 않아도 됩니다.
    - 개인 노트북에 관련 패키지 설치를 희망하는 분들을 위한 매뉴얼입니다.
    - 따라서 설치 오류에 대한 질문을 현장에서는 받지 않겠습니다. (단, 강습회 행사 전날까지는 질문을 받도록 하겠습니다.)


- OS
    - TensorFlow 패키지는 Linux 계열의 운영체제에 적합하지만
    - 사용자 편의를 위해 Windows 10으로 매뉴얼 작성 [^1]


- GPU
    - TensorFlow는 CUDA를 이용해 GPU의 병렬연산 활용 가능 [^2]
    - 본 매뉴얼에서는 편의를 위해 CPU 버전 설치만 다룸
    - GPU 버전 설치 <span style="text-decoration: underline;color: #000080"><a style="color: #000080;text-decoration: underline" href="http://i-systems.github.io/HSE545/machine%20learning%20all/Workshop/PHMAP/python_and_tf_installation_guide_GPU.html" target="_blank">메뉴얼</a></span> (참고용, 본 강습회에서는 다루지 않습니다):

- Anaconda
    - 본 메뉴얼에서는 패키지들 사이의 충돌을 방지하기 위해 Anaconda를 사용하지 않음
    - Anaconda 버전 설치 <span style="text-decoration: underline;color: #000080"><a style="color: #000080;text-decoration: underline" href="http://i-systems.github.io/HSE545/machine%20learning%20all/Workshop/PHMAP/python_and_tf_installation_guide_anaconda.html" target="_blank">메뉴얼</a></span> (참고용, 본 강습회에서는 다루지 않습니다):

[^1]: 2016년 12월, TensorFlow 1.0부터 Windows 7, 10을 지원하나 사용에 일부 제약이 있음. <br> Windows 7, 또는 32 bits PC인 경우 본 매뉴얼 대로 설치하되 에러가 발생하는 부분이 있으면 연락 바람 (Windows 8 미지원)
[^2]: Intel, AMD 계열의 그래픽카드 칩셋은 지원하지 않으며 NVIDA CUDA 3.0 이상만 지원
