## Estimation of homography H

$$ X' = HX $$

where $ X $ and $X'$ are position vectors of key points, and $ H $ is a Perspective Transformation

__Goal: we need to estimate homography $H$ via matching potins between two images__


-----

$$\begin{bmatrix}
    x' \\
    y' \\
    1 \end{bmatrix}
    \sim
    \begin{bmatrix}
    \omega x' \\
    \omega y' \\
    \omega \end{bmatrix}
    =
    \begin{bmatrix}
    \theta_{1} & \theta_{2} & \theta_{3} \\
    \theta_{4} & \theta_{5} & \theta_{6} \\
    \theta_{7} & \theta_{8} & 1 \end{bmatrix} 
    \begin{bmatrix}
    x\\
    y\\
    1 
\end{bmatrix}$$
    
<br>
$$ 
\begin{align*}
x' &= \frac{\theta_1 x+\theta_2 y+\theta_3}{\theta_7 x+\theta_8 y+1} \\
y' &= \frac{\theta_4 x+\theta_5 y+\theta_6}{\theta_7 x+\theta_8 y+1}
\end{align*}
$$
<br>
$$ 
\begin{align*}
\theta_1 x+\theta_2 y+\theta_3 -\theta_7 x'x-\theta_8 x'y-x' &= 0 \\
\theta_4 x+\theta_5 y+\theta_6 -\theta_7 y'x-\theta_8 y'y-y' &= 0 
\end{align*}
$$

<br>
For $m$ pairs of matching potins

- $ \Phi $ is a feature matrix

$$ \Phi = \begin{bmatrix}
            x_{1} & y_{1} & 1 & 0 & 0 & 0 & -x'_{1}x_{1} & -x'_{1}y_{1}\\
            0 & 0 & 0 & x_{1} & y_{1} & 1 & -y'_{1}x_{1} & -y'_{1}y_{1}\\
            \vdots &\vdots &\vdots &\vdots &\vdots &\vdots &\vdots &\vdots\\
            x_{m} & y_{m} & 1 & 0 & 0 & 0 & -x'_{m}x_{m} & -x'_{m}y_{m}\\
            0 & 0 & 0 & x_{m} & y_{m} & 1 & -y'_{m}x_{m} & -y'_{m}y_{m}\end{bmatrix} $$
         
         
           
- $ \theta $ is a column vector for unknown parameters in a perspective transformation $H$



$$ \theta = \begin{bmatrix}
            \theta_{1} \\
            \theta_{2} \\
            \theta_{3} \\
            \theta_{4} \\
            \theta_{5} \\
            \theta_{6} \\
            \theta_{7} \\
            \theta_{8}
            \end{bmatrix} $$
            
- $b$ is a column vector for corresponding positions in the base image

$$ b = \begin{bmatrix}
            x'_{1} \\ 
            y'_{1} \\
            x'_{2} \\
            y'_{2}  \\
            \vdots \\
            x'_{m} \\
            y'_{m} \end{bmatrix} $$
                    
------            


- It ends up becoming a linear regression problem

$$ \min\limits_{\theta} \lVert \Phi\theta - b \rVert _2^2 $$

$$ \theta^* = (\Phi^T\Phi)^{-1}\Phi^T b $$


## Find key points between two images
 
- Suppose these matching points are given.
    - We have manually found the matching points for you, although there is a technique to do this.     


- pos1 and pos2 are matching points between img01 and img02
- pos3 and pos4 are matching points between img02 and img03

```python

pos1 = np.array([[2497, 2016, 2429, 2488, 2580, 2455, 2764, 2694],
                 [870, 709, 1260, 1160, 1236, 1296, 1128, 981]])
pos2 = np.array([[824, 311, 717, 788, 869, 739, 1048, 1000],
                 [849, 584, 1230, 1135, 1218, 1268, 1132, 986]])
pos3 = np.array([[735, 577, 662, 761, 808, 494, 685, 1017],
                 [1129, 1294, 1152, 1268, 1249, 2109, 1088, 1360]])
pos4 = np.array([[2504, 2361, 2435, 2541, 2587, 2402, 2450, 2825],
                 [1210, 1381, 1239, 1347, 1325, 2171, 1175, 1431]])
```

## Visualization of key points

```python
plt.imshow(img01)
plt.plot(pos1[0,:].transpose(),pos1[1,:].transpose(), 'ro')
plt.show()
```

```python
plt.imshow(img02)
plt.plot(pos2[0,:].transpose(),pos2[1,:].transpose(), 'ro')
plt.show()
```

```python
Phi01 = np.zeros((16, 8))
newpos01 = np.zeros((16, 1))

for i in range(0,8):
    Phi01[2*i,:] = np.array([[pos1[0, i], pos1[1, i], 1, 0, 0, 0, -pos2[0, i]*pos1[0, i], -pos2[0, i]*pos1[1,i]]])
    Phi01[2*i+1,:] = np.array([[0, 0, 0, pos1[0, i], pos1[1, i], 1, -pos2[1, i]*pos1[0, i], -pos2[1, i]*pos1[1, i]]])
    
    newpos01[2*i, :] = np.array([[pos2[0, i]]])
    newpos01[2*i+1, :] = np.array([[pos2[1, i]]])
    
    
perspective_theta = LA.inv(Phi01.transpose().dot(Phi01)).dot(Phi01.transpose()).dot(newpos01)
perspective_theta = np.concatenate((perspective_theta, np.ones((1,1))), 0)
perspective_theta = np.reshape(perspective_theta, (3,3))

Phi02 = np.zeros((16, 8))
newpos02 = np.zeros((16, 1))

for i in range(0, 8):
    Phi02[2*i,:] = np.array([[pos3[0, i], pos3[1, i], 1, 0, 0, 0, -pos4[0, i]*pos3[0, i], -pos4[0, i]*pos3[1,i]]])
    Phi02[2*i+1,:] = np.array([[0, 0, 0, pos3[0, i], pos3[1, i], 1, -pos4[1, i]*pos3[0, i], -pos4[1, i]*pos3[1, i]]])
    
    newpos02[2*i, :] = np.array([[pos4[0, i]]])
    newpos02[2*i+1, :] = np.array([[pos4[1, i]]])
    
    
perspective_theta2 = LA.inv(Phi02.transpose().dot(Phi02)).dot(Phi02.transpose()).dot(newpos02)
perspective_theta2 = np.concatenate((perspective_theta2, np.ones((1,1))), 0)
perspective_theta2 = np.reshape(perspective_theta2, (3,3))
```

## Image warping

- Again, do not worry about the image warping (outside lecture's scope)

```python
temp = np.zeros((3,3))
temp[0:2,0:2] = np.eye(2) 

temp[0,2]= 5000.0
temp[1,2]= 2000.0
temp[2,2] = 1.0

new_base_theta = np.dot(temp,np.eye(3))
new_perspective_theta = np.dot(temp,perspective_theta)
new_perspective_theta2 = np.dot(temp,perspective_theta2)

dst1 = cv2.warpPerspective(img01,new_perspective_theta,(12180,5382))
dst2 = cv2.warpPerspective(img02,new_base_theta,(12180,5382))
dst3 = cv2.warpPerspective(img03,new_perspective_theta2,(12180,5382))

(ret,data_map) = cv2.threshold(cv2.cvtColor(dst3, cv2.COLOR_BGR2GRAY),0, 255, cv2.THRESH_BINARY)


plt.imshow(np.maximum(np.maximum(dst1,dst2),dst3))
```
