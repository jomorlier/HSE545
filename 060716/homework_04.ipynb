# Problem 1

The uniform distribution for a continuous variable $x$ is defined by

$$U(x;a,b) = \frac{1}{b-a}, \quad a\leq x \leq b$$

Verify that this distribution is normalized, and find expressions for its mean and variance.

# Problem 2

Consider two urns. The first contains two white and seven black balls, and the second contains five white and six black balls. We flip a fair coin and then draw a ball from the first urn or the second urn depending on whether the outcome was heads or tails. What is the conditional probability that the outcome of the toss was heads given that a white ball was selected?

# Problem 3

Urn 1 contains two white balls and one black ball, while urn 2 contains one white ball and five black balls. One ball is drawn at random from urn 1 and placed in urn 2. A ball is then drawn from urn 2. It happens to be white. What is the probability that the transferred ball was white given this situation?

# Problem 4

Suppose that $p(x,y)$, the joint probability mass function of $X$ and $Y$, is given by

$$p(1,1)=0.5, \quad p(1,2)=0.1, \quad p(2,1) = 0.1, \quad p(2,2)=0.3$$

Calculate the conditional probability mass function of $X$ given that $Y=1$.

# Problem 5

A and B go target shooting together. Both shoot at a target at the same time. Suppose A hits the target with probability 0.7, whereas B, independently, hits the target with probability 0.4. 

(a) Given that exactly one shot hits the target, what is the probability that it was B's shot?


(b) Given that the target is hit, what is the probability that B hits it?

# Problem 6

$$
\begin{align*}
E[x+y] &= E[x]+E[y] \\
\text{var}(x+y) &= \text{var}(x) + \text{var}(y)
\end{align*}$$

# Problem 7

# Problem 8

# Problem 9

Suppose you randomly sample $m$ data points from a normal distribution. You want to estimate the expectation of the distribution. Since the sample was random, the distribution of these points are independent. We can compute the joint distribution by taking the product of the marginal distribution. Suppose that the variance of this normal distribution is known to be 1. Prove that the MLE is the same as a sample mean.
